- 색상: b(파랑), g(초록), c(청록), y(노랑), k(검정), w(흰색)
- 선종류: -(실선), --(파선), -.(일점쇄선), :(점선)
- 마커 모양: o(원), d(다이아), s(사각), ^(삼각), v(역삼각), .(점)

```python
import numpy as np
plt.plot(np.random.randn(10), 'k', label = 'one')
plt.plot(np.random.randn(10)*3, 'r--', label = 'two')
plt.plot(np.random.randn(10)*10, 'g.', label = 'three')

plt.legend()
plt.show()
```

# **다중그래프**

```python
plt.subplot(2,2,1) #2행 2열 중 첫 번째 그래프
plt.plot(np.random.randn(10),'b--')
plt.subplot(2,2,2)
plt.plot(np.random.randn(100),'r',alpha=0.7) #알파 = 불투명도
plt.subplot(2,2,3)
plt.plot(np.random.randn(10),'y^')
plt.subplot(2,2,4)
plt.plot(np.random.randn(10),'g.')
plt.show()
```

# **상자 그래프**
- 연속형 변수의 분포를 살펴볼 수 있음
- 최대값, 최솟값, 중앙값, 사분위수를 표시




1.   제 1사분위수(Q1)와 제 3사분위수(Q3)를 양 끝으로 하는 상자를 그림
2.   상자 안에 중앙값을 수평선으로 표시
3. 상자의 길이, 즉 Q1~Q3까지의 범위를 사분범위라고 하며 이 상자의 길이의 1.5배 안에 있는 데이터 를 기준으로 수염을 표시
4. 수염 바깥에 있는 데이터는 작은 원으로 표현, 이상치로 간주

- Q1(1사분위수): 데이터의 하위 25% 위치에 해당하는 값
- Q3(3사분위수): 데이터의 상위 25% 위치에 해당하는 값


```python
numbers = [0, 1.4, 1.6, 1.8, 1.85, 1.9, 2.2, 2.5, 5.7, 9]
plt.boxplot(numbers)
plt.show()

q1 = np.percentile(numbers, 25)
q3 = np.percentile(numbers, 75)
print(f'q1: {q1},q3: {q3}')
```

# **막대 그래프**
- 범주형 변수끼리 연송형 값을 비교할 수 있음
- plt.bar(x값 리스트, y값 리스트)

```python
subject = ['KOR','ENG','MATH']
grade = [85,76,55]

plt.title("Report")
plt.xlabel('subject')
plt.ylabel('grade')
plt.bar(subject, grade)
plt.ylim(0,100) #y축 범위
plt.text(0,90,'Good!') #x좌표가 0이고 y좌표가 90인 위치에 'Good!' 텍스트 추가
plt.show()
```

# **파이차트**
- 범주형 변수끼리 연속형 값의 비율을 비교할 수 있음

```python
import pandas as pd
example_series = pd.Series([20, 15, 5, 5, 10], index = ['Apple','Banana','Cherry','Kiwi','Grape'])
import matplotlib.pyplot as plt
plt.pie(example_series, labels = example_series.index, autopct = '%.1f%%')
plt.show()
```

# **네트워크 그래프**

```python
import networkx as nx

G = nx.DiGraph()
G.add_nodes_from(['A','B','C','D','E'])
G.add_edges_from([('A','B'),('B','C'),('C','D')])

nx.draw(G, with_labels = True)
```

```python
import networkx as nx

g1 = nx.Graph()
g1.add_node('David')
g1.add_node('John')
g1.add_node('Mary')
g1.add_node('Julia')
g1.add_node('Juda')
g1.add_nodes_from(['Judy','Karen'])
g1.remove_node('Julia')

g1.add_edge('David','John')
g1.add_edge('David','Mary')
g1.add_edge('David','Karen')
g1.add_edges_from([('John','Mary'),('John','Judy')])
g1.remove_edge('Mary','John')

nx.draw(g1, with_labels = True, font_weight = 'bold')
```

```python
d = dict(g1.degree)
nx.draw(g1, nodelist=d.keys(),node_size = [v*1000 for v in d.values()], with_labels = True, font_weight = 'bold')
g1.nodes
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('기온.csv', encoding='cp949')
df.head(3)
```

```python
df.info()
```

```python
df2 = df.fillna(method='ffill')
df2.info()
```

```python
df2.rename(columns={'최저기온(°C)':'min_temp'}, inplace=True)
df2.rename(columns={'평균기온(°C)':'avg_temp'}, inplace=True)
df2.rename(columns={'최고기온(°C)':'max_temp'}, inplace=True)
df2.head(3)
```

```python
plt.rc('font',family = 'NanumGothic')
plt.rcParams['axes.unicode_minus'] = False

plt.title('서울시 2022년도 기온 변화')
plt.plot(range(1,len(df2)+1),df2['max_temp'],label='최고기온',c='r')
plt.plot(range(1,len(df2)+1),df2['avg_temp'],label='평균기온',c='y')
plt.plot(range(1,len(df2)+1),df2['min_temp'],label='최저기온',c='b')
plt.xlabel('월')
plt.ylabel('기온')
plt.legend()
plt.show()
```

```python
df2['일시'] = pd.to_datetime(df2['일시'],format = '%Y-%m-%d')
df2['일시']
```

```python
df3 = df2.set_index('일시')
df3.head(3)
```

```python
df_mar = df3[pd.DatetimeIndex(df3.index).month == 3]
```

```python
plt.title('서울시 2022년 3월 기온 변화')
plt.plot(range(1,32),df_mar['max_temp'],label='최고기온',c='r')
plt.plot(range(1,32),df_mar['avg_temp'],label='평균기온',c='y')
plt.plot(range(1,32),df_mar['min_temp'],label='최저기온',c='b')

plt.xlabel('3월')
plt.ylabel('기온')
plt.xlim(1, 31)
plt.ylim(-7, 22)
plt.legend()
plt.rcParams['figure.figsize'] = (12, 8)

plt.show()
plt.savefig('서울.png')
```
