# **데이터프레임: 판다스의 기본 구조인 자료구조 객체**

1.   시리즈 여러 개를 묶어서 데이터 프레임을 만들어 2차원 배열과 비슷

- 값(데이터): 행과 열이 교차하는 곳에 지정되는 데이터

# **배열복사**
- 얕은복사: 등호를 이용하는 복제, 원본배열 수정하면 복사본이 함께 바뀜 그 반대도 마찬가지
- 깊은복사: copy함수를 사용 원본에 영향 x

```python
import numpy as np
a = np.arange(6)
b = a #얕은 복사

print(a)
print(b is a)

b[0] = 10
print(a)
```

```python
a = np.arange(6)
c = a.copy() #깊은 복사

c[0] = 20
print('A: ',a)
print('C: ',c)
```

```python
import pandas as pd

list2 = [['허준호','남자',30,183],
               ['이가원','여자',24,162],
               ['배규민','남자',23,179],
               ['고고림','남자',21,182],
               ['이새봄','여자',28,160],
               ['이보람','여자',26,163],
               ['이루리','여자',24,157],
               ['오다현','여자',24,172]]
col_names = ['이름','성별','나이','키']

df = pd.DataFrame(list2, columns=col_names)
print(df)
```

```python
df[df['키'] > 180].head() #''로 인하여 자료형이 문자열이기 때문에 숫자 데이터는 비교조건으로 할 때 자료형을 숫자형으로 바꿔줘야함 안그럼 오류 ㅈㄴ 뜸 ㅅㄱ
```

```python
df[df['나이'].isin([21,23])] #조건식에 isin()함수, and기호, or기호, str.contains()함수를 사용하여 더 복잡한 조건으로 조회할 수 도 있음
```

```python
df[(df['성별'] == '여자') & (df['키']>160)]
```

```python
df[(df['나이'] >= 28) | (df['성별'] == '남자')]
```

```python
df[df['이름'].str.contains('봄')]
```
